```python
# Importa as funções necessárias do Flask
from flask import Flask, jsonify, request

# Cria a aplicação Flask
app = Flask(__name__)

# Lista de produtos (simulando um banco de dados)
produtos = [
    {
        "id": 1,
        "nome": "Camiseta",
        "preco": 50.00
    },
    {
        "id": 2,
        "nome": "Tênis",
        "preco": 120.00
    }
]

# Rota para listar todos os produtos
# Método: GET
@app.route('/produtos', methods=['GET'])
def listar_produtos():
    # Retorna a lista de produtos em formato JSON
    return jsonify(produtos)


# Rota para adicionar um novo produto
# Método: POST
@app.route('/produtos', methods=['POST'])
def adicionar_produto():
    # Recebe os dados enviados no corpo da requisição
    novo_produto = request.get_json()

    # Adiciona o novo produto à lista
    produtos.append(novo_produto)

    # Retorna uma mensagem de sucesso e o código 201 (criado)
    return jsonify({
        "mensagem": "Produto adicionado com sucesso!"
    }), 201


# Rota para atualizar um produto existente
# Método: PUT
@app.route('/produtos/<int:id>', methods=['PUT'])
def atualizar_produto(id):
    # Procura o produto pelo ID
    produto = next((p for p in produtos if p['id'] == id), None)

    # Verifica se o produto foi encontrado
    if produto is None:
        # Retorna erro 404 caso não exista
        return jsonify({
            "erro": "Produto não encontrado!"
        }), 404

    # Recebe os novos dados enviados
    dados_atualizados = request.get_json()

    # Atualiza os dados do produto encontrado
    produto.update(dados_atualizados)

    # Retorna mensagem de sucesso
    return jsonify({
        "mensagem": "Produto atualizado com sucesso!"
    })


# Rota para remover um produto
# Método: DELETE
@app.route('/produtos/<int:id>', methods=['DELETE'])
def remover_produto(id):
    # Procura o produto pelo ID
    produto = next((p for p in produtos if p['id'] == id), None)

    # Verifica se o produto existe
    if produto is None:
        # Retorna erro 404 se não encontrar
        return jsonify({
            "erro": "Produto não encontrado!"
        }), 404

    # Remove o produto da lista
    produtos.remove(produto)

    # Retorna mensagem de sucesso
    return jsonify({
        "mensagem": "Produto removido com sucesso!"
    })


# Executa a aplicação
if __name__ == '__main__':
    # Inicia o servidor Flask em modo de depuração
    app.run(debug=True)
```
